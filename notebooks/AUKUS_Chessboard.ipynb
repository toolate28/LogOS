```python
# @title 0. Setup · imports & constants
import subprocess, sys
subprocess.run([sys.executable, "-m", "pip", "install", "-q", "pandas"], check=False)

import json, hashlib, datetime as dt
from pathlib import Path
from dataclasses import dataclass, field, asdict
from typing import Optional
import pandas as pd

NOTEBOOK_DIR = Path.cwd() / "notebooks" if (Path.cwd() / "notebooks").exists() else Path.cwd()
MOVES_PATH = NOTEBOOK_DIR / "aukus_moves.jsonl"
NEGATIVE_SPACE_PATH = NOTEBOOK_DIR / "negative_space.json"

ALPHA, OMEGA, CONSERVATION = 7, 8, 15
EPSILON = 0.00055
WAVE_THRESHOLD = 0.85
K22 = {"vertices": 22, "edges": 41}

CORRIDORS = ["A_bauxite_bilbao", "B_msb_megawatt", "C_quasicrystal", "D_agent_aukus", "E_night_near"]
RANKS = ["space", "materials", "power", "regulatory", "compute", "cognitive"]
FILES = ["alpha_lane", "omega_lane", "witness_lane", "srac_lane"]
TERRITORIES = ["AU", "US", "UK"]

print(f"[INIT] α+ω={CONSERVATION} ε={EPSILON} K22={K22['vertices']}v{K22['edges']}e")
print(f"[INIT] Moves log: {MOVES_PATH}")
```

```python
# @title 1. Move schema · append conversation telemetry

@dataclass
class ChessMove:
    date: str
    move_id: str
    rank: str          # space | materials | power | regulatory | compute | cognitive
    file: str          # alpha_lane | omega_lane | witness_lane | srac_lane
    territory: str     # AU | US | UK
    corridor: str        # A..E
    action: str
    insight: str
    wave: float
    void_delta: float  # negative space shrinkage estimate (-1..+1)
    obstruction: str   # Serre-Scarr page stuck (or "lifted")
    source: str = "conversation"

def append_move(m: ChessMove) -> None:
    MOVES_PATH.parent.mkdir(parents=True, exist_ok=True)
    with MOVES_PATH.open("a", encoding="utf-8") as f:
        f.write(json.dumps(asdict(m), ensure_ascii=False) + "\n")
    print(f"[MOVE] {m.move_id} · {m.territory}/{m.rank} · WAVE={m.wave:.2f}")

def load_moves() -> pd.DataFrame:
    if not MOVES_PATH.exists():
        return pd.DataFrame()
    rows = [json.loads(line) for line in MOVES_PATH.read_text(encoding="utf-8").splitlines() if line.strip()]
    return pd.DataFrame(rows)

def invariant_ok(wave: float) -> bool:
    return wave >= WAVE_THRESHOLD

print("[SCHEMA] ChessMove ready. Use append_move() after each session.")
```

```python
# @title 3. Chessboard heatmap · rank × territory

if len(df_moves) == 0:
    print("[BOARD] No moves yet — run seed cell")
else:
    heat = df_moves.groupby(["rank", "territory"]).agg(
        moves=("move_id", "count"),
        mean_wave=("wave", "mean"),
        void_sum=("void_delta", "sum"),
    ).reset_index()
    pivot = heat.pivot(index="rank", columns="territory", values="mean_wave").reindex(RANKS)
    print("=== Mean WAVE by rank × territory ===")
    display(pivot)
    print("\n=== Corridor activity ===")
    display(df_moves.groupby("corridor")["move_id"].count().sort_values(ascending=False))
```
